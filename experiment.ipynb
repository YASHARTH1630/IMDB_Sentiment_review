```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout,Dense,Embedding,SimpleRNN
```

```python
##load the dataset
max_features=10000 ##vocabulary size of each word
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=max_features)

```

```python
X_train[0] ##predefine encoding
```

```python
word_index=imdb.get_word_index()

reversed_word_index=dict([(value,key) for (key,value) in word_index.items()])

decoded_review=' '.join([reversed_word_index.get(i-3,'?') for i in X_train[0]])
decoded_review
```

```python
maxlen=500 ##padding causes trimming of data into 500 length of array



X_train=sequence.pad_sequences(X_train,maxlen=maxlen)
X_test=sequence.pad_sequences(X_test,maxlen=maxlen)
```

```python
X_train.shape
```

```python
model=Sequential() ##128 is dimension of word after embedding and for every review there would be maxlen length 
maxlen=500
model.add(Embedding(max_features,128,input_shape=(maxlen,)))
model.add(SimpleRNN(128,dropout=0.2,recurrent_dropout=0.2,activation='leaky_relu'))

model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
## Create an instance of EarlyStoppping Callback
from tensorflow.keras.callbacks import EarlyStopping
earlystopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
earlystopping
```

```python
history=model.fit(
    X_train,y_train,epochs=15,batch_size=32,
    validation_split=0.2,
    callbacks=[earlystopping]
)
```

```python
model.save("simple_run_imdb.h5")
```
